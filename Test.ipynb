```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Example data
np.random.seed(42)
y_data = np.linspace(0, 10, 20)  # Y-coordinates
x_data = [np.random.normal(loc=y, scale=0.5, size=100) for y in y_data]  # X-values for density

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

violin_height = 3  # Height of each violin plot

# Plot each horizontal half-violin
for y, x_vals in zip(y_data, x_data):
    # Calculate the density
    kde = gaussian_kde(x_vals)
    x_range = np.linspace(min(x_vals) - 1, max(x_vals) + 1, 200)  # Range for density plot
    density = kde(x_range)
    
    # Normalize the density to scale to the desired height
    density /= density.max()  # Normalize the density to [0, 1]
    density *= violin_height  # Scale density to match desired height
    
    # Draw the half-violin symmetrically centered at the y-coordinate
    y_lower = y - violin_height / 2  # Bottom edge of the violin
    y_upper = y + violin_height / 2  # Top edge of the violin
    ax.fill_between(x=x_range, y1=y_lower, y2=y_lower+density,
                     color="blue", alpha=0.5)
    
    # Add a scatter point for clarity
    ax.scatter([np.mean(x_vals)], [y], color='black', zorder=3)

# Customize the plot
ax.set_xlabel("X-values")
ax.set_ylabel("Y-values")
ax.set_title("2D Scatter Plot with Correctly Centered Horizontal Half-Violin Plots")

# Set proper limits for better visualization
ax.set_xlim(-2, 12)
ax.set_ylim(-1, 11)

plt.show()
```
